```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
from time import sleep
```

```python
def f(theta,xi):
    return -(g/l)*np.sin(theta)
```

```python
h=0.01;
g=9.81;
l=0.5;
theta=np.pi/5;
xi=0.0;
n=1000
ls=n*[l]
```

```python
def Euler(t,theta,xi,h):
    xi=xi+f(theta,xi)*h
    theta=theta+xi*h
    x=l*np.cos(theta)
    y=-l*np.sin(theta)
    return x,y,theta,xi,t*h
```

```python
xs=[];ys=[];thetas=[];xis=[];times=[]
for t in range(n):
    x, y, theta, xi, time = Euler(t, theta, xi, h)
    xs.append(x); ys.append(y);
    thetas.append(theta);xis.append(xi)
    times.append(time)
```

```python
for i in range(1,len(xis),5):
    sleep(1)
    clear_output(wait=True)
    fig = plt.figure(figsize=(12,3))
    ax1 = fig.add_subplot(1,3,2)
    ax2 = fig.add_subplot(1,3,3)
    ax0 = fig.add_subplot(1, 3, 1, projection='polar') # AÃ±adir subplot polar
    ax0.plot(thetas[:i], ls[:i], marker="x", markersize=5, color="red") # Graficar datos en el subplot polar
    ax1.scatter(times[:i], thetas[:i], color="black")
    ax1.set_ylim(-1,1)
    ax1.set_xlim(0,2*np.pi)
    ax1.grid()
    ax2.scatter(thetas[:i], xis[:i], color="purple")
    ax2.set_xlim(-4,4)
    ax2.set_ylim(-4,4)
    ax2.grid()
    plt.draw()
    plt.show()
    
```
