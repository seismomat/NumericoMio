$$\frac{d^2 \theta}{dt^2}= - \frac{g}{l}sen(\theta) - \beta \frac{d \theta}{dt}$$

$$ v_{n+1}=v_{n}- (\frac{g}{l}sen(\theta_ n) +\beta v_n) $$
$$ \theta_{n+1}=\theta_{n}+hv_n $$

Runge Kutta de segundo orden 

$$ k_1=hv(x_n,t_n)$$
$$ l_1= h f(x_n,v_n,t_n)$$
$$ k_2=h(v_n+l_1)$$
$$ l_2= h f(x_n +k_1,v_n+l_1,t_{n+1})$$

$$x_{n+1}= x_n + \frac{1}{2}(k_1+k_2) $$
$$ v_{n+1}= v_n +\frac{1}{2}(l_1+l_2)$$

```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
from time import sleep
```

```python
def f(theta,xi):
    return -(g/l)*np.sin(theta)-0.6*xi
```

```python
h=0.01;
g=9.81;
l=0.5;
theta=np.pi/5;
xi=0.0;
n=1000
ls=n*[l]
```

```python
def RK2(t,theta,xi,h):
    k1=h*xi
    l1=f(theta,xi)*h
    k2=h*(xi+l1)
    l2=f(theta+k1,xi+l1)*h
    xi=xi+(l1+l2)/2
    theta=theta+(k1+k2)/2
    x=l*np.cos(theta)
    y=-l*np.sin(theta)
    return x,y,theta,xi,t*h
```

```python
xs=[];ys=[];thetas=[];xis=[];times=[]
for t in range(n):
    x, y, theta, xi, time = RK2(t, theta, xi, h)
    xs.append(x); ys.append(y);
    thetas.append(theta);xis.append(xi)
    times.append(time)
```

```python
for i in range(1,len(xis),5):
    sleep(1)
    clear_output(wait=True)
    fig = plt.figure(figsize=(12,3))
    ax1 = fig.add_subplot(1,3,2)
    ax2 = fig.add_subplot(1,3,3)
    ax0 = fig.add_subplot(1, 3, 1, projection='polar') # Aadir subplot polar
    ax0.plot(thetas[:i], ls[:i], marker="x", markersize=5, color="red") # Graficar datos en el subplot polar
    ax1.scatter(times[:i], thetas[:i], color="black")
    ax1.set_ylim(-1,1)
    ax1.set_xlim(0,2*np.pi)
    ax1.grid()
    ax2.scatter(thetas[:i], xis[:i], color="purple")
    ax2.set_xlim(-1,1)
    ax2.set_ylim(-3,3)
    ax2.grid()
    plt.draw()
    plt.show()
```
