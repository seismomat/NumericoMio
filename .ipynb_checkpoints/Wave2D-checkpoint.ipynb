```python
import numpy as np
import numba as nb
import matplotlib.pyplot as plt
from IPython.display import clear_output
from time import sleep

# Parámetros
Lx = 1.0     # Longitud del dominio en x
Ly = 1.0     # Longitud del dominio en y
T = 1.0      # Tiempo total
c = 1.0      # Velocidad de la onda
dx = 0.01    # Tamaño de paso espacial en x
dy = 0.01    # Tamaño de paso espacial en y
dt = 0.01    # Tamaño de paso temporal
# Calculando el número de puntos en el espacio y en el tiempo
nx = int(Lx / dx) + 1
ny = int(Ly / dy) + 1
nt = int(T / dt) + 1
# Condiciones iniciales
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)
X, Y = np.meshgrid(x, y)
```

```python
def simulation(T, dt,u,x,y):
    nt = int(T / dt) + 1
    for t in range(1, nt):           
        clear_output(wait=True)
        # Graficar la solución
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        X, Y = np.meshgrid(x, y)
        ax.plot_surface(X, Y, u[t], cmap='hot')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Amplitud')
        ax.set_title('Ecuación de onda en 2D con diferencias finitas')
        plt.show()

    return 
```

```python
@nb.jit(nopython=True)
def solve_wave_equation_2d(Lx,Ly, T, c, dx,dy, dt,X,Y):
    # Calculando el número de puntos en el espacio y en el tiempo
    nx = int(Lx / dx) + 1
    ny = int(Ly / dy) + 1
    nt = int(T / dt) + 1

    # Inicialización de la matriz u
    u = np.zeros((nt, nx, ny))

    u[0] = np.sin(2 * np.pi * X) * np.sin(2 * np.pi * Y)

    # Iteración temporal
    for t in range(1, nt):
        for i in range(1, nx-1):
            for j in range(1, ny-1):
                u[t, i, j] = 2*u[t-1,i,j]- u[t-2,i, j] + \
                (c*dt/dx)**2*(u[t-1,i+1,j] + u[t-1,i-1,j]- 2*u[t-1,i,j]) + \
                (c*dt/dy)**2*(u[t-1,i,j+1]+u[t-1,i,j-1] - 2*u[t-1,i,j])

    return u
```

```python
U=solve_wave_equation_2d(Lx,Ly, T, c, dx,dy, dt,X,Y)
simulation(T, dt,U,x,y)
```
