```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
L = 1.0  # Longitud del dominio
n = 10   # Número de elementos
dx = L / n  # Tamaño de cada elemento

# Definición de la función de carga f(x)
def f(x):
    return np.sin(np.pi * x)

# Discretización del dominio
x = np.linspace(0, L, n+1)  # Puntos nodales

# Ensamblaje del sistema lineal
A = np.zeros((n+1, n+1))  # Matriz de coeficientes
b = np.zeros(n+1)  # Vector de carga

for i in range(n):
    # Elemento i
    x_left = x[i]
    x_right = x[i+1]
    
    # Longitud del elemento
    h = x_right - x_left
    
    # Contribución del elemento al sistema lineal
    A[i, i] += 1 / h
    A[i, i+1] -= 1 / h
    A[i+1, i] -= 1 / h
    A[i+1, i+1] += 1 / h
    
    b[i] += dx / 2 * f(x_left)
    b[i+1] += dx / 2 * f(x_right)

# Condiciones de contorno
A[0, 0] = 1.0
A[-1, -1] = 1.0
b[0] = 0.0
b[-1] = 0.0

# Resolución del sistema lineal
u = np.linalg.solve(A, b)

# Visualización de la solución
plt.plot(x, u)
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Solución de la ecuación de Poisson en 1D')
plt.grid(True)
plt.show()
```
